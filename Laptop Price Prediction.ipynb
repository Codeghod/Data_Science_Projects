```python
df  = df.join(pd.get_dummies(df['TypeName']))
df = df.drop('TypeName', axis = 1)
```

```python
df.head()
```

```python
df['ScreenResolution'] = df['ScreenResolution'].str.split(' ').apply(lambda x: x[-1])
df['ScreenWidth'] = df['ScreenResolution'].str.split('x').apply(lambda x: x[0]).astype('int')
df['ScreenLength'] = df['ScreenResolution'].str.split('x').apply(lambda x: x[1]).astype('int')
```

```python
df.drop('ScreenResolution', axis = 1, inplace = True)
```

```python
df['Cpu_brand'] = df['Cpu'].str.split(' ').apply(lambda x: x[0])
df['Cpu_speed'] = df['Cpu'].str.split(' ').apply(lambda x: x[-1])
```

```python
df.drop('Cpu', axis = 1, inplace = True)
```

```python
df['Cpu_speed'] = df['Cpu_speed'].str[:-3]
```

```python
df['Cpu_speed'] = df['Cpu_speed'].astype(float)
```

```python
df.head()
```

```python
df['Ram'] = df['Ram'].str[:1].astype('int')
```

```python
df['Memory_amount'] = df['Memory'].str.split(' ').apply(lambda x: x[0])
df['Memory_type'] = df['Memory'].str.split(' ').apply(lambda x: x[1])
```

```python
def turn_memory_into_MB(value):
    if 'GB' in value:
        return float(value[:value.find('GB')]) * 1000
    elif 'TB' in value:
        return float(value[:value.find('TB')]) * 1000000
```

```python
df['Memory_amount'] = df['Memory_amount'].apply(turn_memory_into_MB)
```

```python
df = df.join(pd.get_dummies(df['Memory_type']))
```

```python
df = df.drop('Memory_type', axis = 1)
```

```python
df = df.drop('Memory', axis = 1)
```

```python
df.head()
```

```python
df['Weight'] = df['Weight'].str[:-2].astype(float)
```

```python
df["GPU Brand"] = df['Gpu'].str.split(' ').str[0]
```

```python
df.drop('Gpu', axis = 1, inplace  = True)
```

```python
df = df.join(pd.get_dummies(df['OpSys']))
df.drop('OpSys', axis = 1, inplace = True)
```

```python
df.head()
```

```python
# df.join(pd.get_dummies(df['Cpu_brand'], prefix = 'Cpu')

Cpu_categories = pd.get_dummies(df['Cpu_brand'])
Cpu_categories.columns = [col + '_Cpu' for col in Cpu_categories.columns]

df = df.join(Cpu_categories)
df = df.drop('Cpu_brand', axis = 1)
```

```python
Gpu_categories = pd.get_dummies(df['GPU Brand'])
Gpu_categories.columns = [col + '_Gpu' for col in Gpu_categories.columns]

df = df.join(Gpu_categories)
df = df.drop('GPU Brand', axis = 1)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.model_selection import cross_val_score
```

```python
X = new_df.drop('Price_euros', axis = 1)
y = new_df['Price_euros']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.10, random_state = 5)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
forest = RandomForestRegressor()
forest.fit(X_train_scaled, y_train)
```

```python
forest.score(X_test_scaled, y_test)
```

## XGBOOST REGRESSOR

```python
from xgboost import XGBRegressor

x_reg = XGBRegressor()

x_reg.fit(X_train_scaled, y_train)
```

```python
x_reg.score(X_test_scaled, y_test)
```

## Cross Validation

```python
# Cross validation on RandomForest Algorithm

cvs = cross_val_score(RandomForestRegressor(), X, y, cv = 5)
```

```python
print(cvs)
```

```python
np.mean(cvs)
```

```python
# Cross validation on XGBoost Algorithm

cxs = cross_val_score(XGBRegressor(), X, y, cv = 5)
```
